```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data Preprocessing

```
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)]) # apriori model expects elements in the list must be strings
```

## Training the Apriori model on the dataset

```
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2) # (3*7)/7501 for min_support ie minimum of 3 common transactions per day,
# 7 days a week, default value for min_confidence is 0.8... so you can keep on dividing it by 2 to get a good number of rules,
# a good min_lift is a minimum of 3 onwards, min_length and max_length are set to 2 because of our business requirements which is representing buy one product A get another product B (hence we want a good association between two products)
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```
results = list(rules)
results
```

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
